```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('//Users/trsaivarun/Desktop/git_uploads/DataAnalysis_/london_bike_sharing/london_merged.csv')
```

```python
df.head()
```

```python
new_cols={
    'timestamp':'timestamp',
    'cnt':'count_bikes',
    't1':'temp_real_C',
    't2':'temp_feelslike',
    'hum':'humidity_per',
    'wind_speed':'wind_speed_kph',
    'weather_code':'weather',
    'is_holiday':'is_holiday',
    'is_weekend':'is_weekend',
    'season':'season'
}

df.rename(new_cols,axis=1,inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df['season']=df['season'].astype(int)
df['is_weekend']=df['is_weekend'].astype(int)
df['is_holiday']=df['is_holiday'].astype(int)
df['weather']=df['weather'].astype(int)
```

```python
df.head()
```

```python
weather_code = {
    1:'Clear',
    2:'scattered clouds',
    3:'Broken clouds',
    4:'Cloudy',
    7:'Rain',
    10:'Thunderstorm',
    26:'Snowfall',
    94:'Freezing Fog'
}
```

```python
df['weather']=df['weather'].map(weather_code)
```

```python
 # 0-spring ; 1-summer; 2-fall; 3-winter
```

```python
season={
    0:'spring',
    1:'summer',
    3:'winter'
}
```

```python
df['season']=df['season'].map(season)
```
